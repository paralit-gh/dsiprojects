# Introduction and Problem Statement

**Introduction**

The video game industry has performed tremendously well in recent times - sales of video games in the U.S. reached a staggering USD 56.9 billion in 2020, representing a 27% increase over 2019. Along with the rise of interest in e-sports, the COVID-19 pandemic also indirectly fueled this growth, as more and more people began to seek forms of entertainment while confining themselves at home to seek refuge from the coronovirus.

However, the video game market is a highly competitive one, and video gamers are often spoilt for choice as developers and publishers alike compete intensely to create and publish the next big hit. In an extreme example, the free-to-play cross-platform game 'Genshin Impact' released by miHoYo in September 2020 - which quickly exploded in popularity worldwide - earned approximately $393 million in revenue within a mere 2 months after release.

---

**Business Problem**

As a data scientist working in a data analytics consultancy, we have been tasked by our client, a large and renowned video game publisher to help them better **discern the sentiment of their users' reviews**, to increase sales revenue and decrease the churn rate. As review scores of games on digital distribution systems such as Steam and Epic Games are often correlated to the sales revenue, it is of paramount importance to understand the user base's opinions, so as to work with the relevant game developers to quickly address any player concerns. These could include issues with the gameplay, graphics, audio, community (hackers/cheaters) or poor game performance on particular machines, among others. 

At the same time, the publisher is also keen to find out the game elements that modern video gamers enjoy, in order to work towards producing higher quality, well-received games.

---

**Problem Statement**

For three popular video games of *different genres* (party game/battle royale shooter/open world action RPG),
- To **build a classification model** and to predict if a particular game review is a *positive* or *negative* one
    - To investigate which are the top predictive text features to determine if a review is *positive* or *negative*
    - Models to be scored on their **ROC-AUC score** and **accuracy**
- To **perform topic modelling** of the various reviews, segregated by their actual sentiment
    - *Positive reviews* - understand what people enjoy, to release future content/expansions for existing games or develop new games that recreate these aspects
    - *Negative reviews* - understand what aspects to improve on, release patches/bug fixes to address these issues in order for player retention
    - Attempt topic classification of reviews to find out their dominant topic

**Evaluation metrics**
- **Accuracy** provides an easily understood metric of the proportion of classifications that were done correctly

$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$

- **ROC-AUC** quantifies the model's ability to separate the positive and negative class distributions. It can be a useful metric when classes are imbalanced and we care both about correctly classifying the negative class and positive class 
    - As a game developer/publisher, it may not be ideal to focus only on sieving out negative reviews - there should be equal focus on identifying positive reviews to understand what players enjoy, and what they should deliver more of
- **Topic coherence score** is used to measure the quality of the topic model

## Filter out non-English reviews

Although we had selected 'English' as our language of choice while performing the scraping, there were several non-English reviews in our resultant datasets. It is likely they were tagged erroneously by users during submission. As such, we will employ the `langdetect` library to filter these out.

```python
# create function to detect language, or return None if error is surfaced

def apply_detect_lang(text):
    try:
        return detect(text)
    except:
        return None
```

```python
# use langdetect library to detect the language of the review

for df_name, df in dfs.items():
    print('-------')
    print(f'Determining actual language of reviews for {df_name}...')
    df['actual_language'] = df['review'].apply(apply_detect_lang)
    print(f'Actual language determined.')
    print('-------')          

    print(f'Number of rows before dropping non-English reviews: {df.shape[0]}')
    
    no_rows_initial = df.shape[0]
    
    df.drop(df[~(df['actual_language'] == 'en')].index, inplace=True)
    print(f'Number of rows after dropping non-English reviews: {df.shape[0]}')
    
    no_rows_final = df.shape[0]
    
    print(f'Rows dropped: {no_rows_initial - no_rows_final}')
    print(' ')
```

## Clean text with `regex`

```python
# create function to strip our text of any hyperlinks and other special characters (excluding ') using regex

def clean_text(text):
    """
    This function takes in a text string and removes any http hyperlinks and special characters such as [] tags, 
    non-English characters and returns the processed string. 
    """
    
    # search the string for any http/https hyperlinks and replaces with blank space
    stripped_html_text = re.sub(r"http\S+",
                                ' ',
                                text)
    
    # search the string for any steam markup tags and replaces with blank space
    stripped_markup_text = re.sub(r"\[.*?\]",
                                ' ',
                                stripped_html_text)
    
    # search the string for any remaining characters not in the given ASCII range (0 - 122)
    
    stripped_special_text = re.sub(r"([^\x00-\x7F])+",
                                   ' ',
                                   stripped_markup_text)
    
    # instantiate regex Tokenizer to match only word characters and apostrophes and tokenize string
    tokenizer = RegexpTokenizer(r"[\w']+")
    text_token = tokenizer.tokenize(stripped_special_text.lower())
    
    # return the joined string
    return (' '.join(text_token))
```

```python
# create function to strip our text of any hyperlinks and other special characters (excluding ') using regex
# retain punctuation and capitalization, to be used for Lexicon-based sentiment analysis

def clean_text_with_punctuation(text):
    """
    This function takes in a text string and removes any http hyperlinks and special characters such as [] tags, 
    non-English characters and returns the processed string with intact punctuation and capitalization.
    """
    
    # search the string for any http/https hyperlinks and replaces with blank space
    stripped_html_text = re.sub(r"http\S+",
                                ' ',
                                text)
    
    # search the string for any steam markup tags and replaces with blank space
    stripped_markup_text = re.sub(r"\[.*?\]",
                                ' ',
                                stripped_html_text)
    
    # search the string for any remaining characters not in the given ASCII range (0 - 122) - e.g. Chinese, Russian
    
    stripped_special_text = re.sub(r"([^\x00-\x7F])+",
                                   ' ',
                                   stripped_markup_text)
    
    # instantiate regex Tokenizer to match only word characters and apostrophes and tokenize string
    tokenizer = RegexpTokenizer(r"[\w'.!]+")
    text_token = tokenizer.tokenize(stripped_special_text)
    
    # return the joined string
    return (' '.join(text_token))
```

```python
# apply clean_text function to datasets

for df in dfs.values():
    df['clean_text'] = df['review'].apply(clean_text)
```

```python
# apply clean_text_with_punctuation function to datasets

for df in dfs.values():
    df['clean_text_punct'] = df['review'].apply(clean_text_with_punctuation)
```

```python
# drop rows that have empty clean_text feature

for df in dfs.values():
    df.drop(df[~df['clean_text'].astype('bool')].index, inplace=True)
```

## Encoding of target value

Steam reviews are tagged as either 'Recommended' or 'Not Recommended'. We will translate this rating to mean 'Positive' or 'Negative' respectively and encode the target variable accordingly (Positive - 1, Negative - 0).

```python
# voted_up - True means positive, False means negative

def encode_target_variable(df):
    df.rename(columns={'voted_up': 'label'}, inplace=True)
    df['label'] = df['label'].astype(int)
```

```python
# encode target variable for each dataframe into 0 and 1

for df in dfs.values():
    encode_target_variable(df)
```

```python
# define a function to plot histograms, and the mean and median values

def plot_hist(dataframe, col_name, titlename, bins=30):
    """
    Plots a histogram for the distribution of a dataframe column.
    """
    fig, ax = plt.subplots(figsize=(12,8))
    sns.histplot(dataframe, x=col_name, discrete=True, bins=bins, ax=ax)
    plt.xlabel(col_name)
    plt.ylabel('Count')
    plt.xlim(0, np.percentile(dataframe[col_name], 95))
    plt.title(titlename, fontsize=20)
    
    # plot the mean and median
    plt.axvline(dataframe[col_name].mean(), color='deepskyblue', linestyle='solid', linewidth=1) 
    plt.axvline(dataframe[col_name].median(), color='midnightblue', linestyle='dotted', linewidth=1)
    plt.text(0.90, 0.95, 'Mean: {:.1f}'.format(dataframe[col_name].mean()), color='deepskyblue', transform=ax.transAxes)
    plt.text(0.90, 0.90, 'Median: {:.1f}'.format(dataframe[col_name].median()), color='midnightblue', transform=ax.transAxes)
```

```python
# create a long string for each dataset to be passed into the wordcloud

pubg_cloud =  " ".join([text for text in pubg_df['clean_text']])
cyberpunk_cloud = " ".join([text for text in cyberpunk_df['clean_text']])
fallguys_cloud = " ".join([text for text in fallguys_df['clean_text']])
```

```python
# update our stopwords list with common words such as 'play', 'game', 'good' 

stopwords = set(STOPWORDS)
stopwords.update(['game', 'play', 'good'])
```

```python
# create word cloud for pubg_df

pubg_wc = WordCloud(width=1000, height=1000, max_words=200, background_color='white', stopwords=stopwords, 
                      random_state=1, colormap='BrBG').generate(pubg_cloud)
```

```python
# display word cloud for pubg_df

plt.figure(figsize=(15,15))
plt.imshow(pubg_wc)
plt.axis("off");
```

'Fun', 'friend' make prominent appearances in the word cloud. However, there are several negative words such as 'hacker', 'bug', 'lag', 'cheater' as well.

There are not many words relating to the gameplay itself.

```python
# create word cloud for cyberpunk_df

cyberpunk_wc = WordCloud(width=1000, height=1000, max_words=200, background_color='white', stopwords=stopwords, 
                      random_state=1, colormap='RdPu').generate(cyberpunk_cloud)
```

```python
# display word cloud for cyberpunk_df

plt.figure(figsize=(15,15))
plt.imshow(cyberpunk_wc)
plt.axis("off");
```

Cyberpunk is not a multiplayer game - hence the notable absence of 'friend' (as well as 'fun') compared to the other two.

The largest word is 'bug' - which is indicative of the many problems that plagued the game's release, rendering the game unplayable for many.

```python
# create word cloud for fallguys_df

fallguys_wc = WordCloud(width=1000, height=1000, max_words=200, background_color='white', stopwords=stopwords, 
                      random_state=1, colormap='winter').generate(fallguys_cloud)
```

```python
# display word cloud for fallguys_df

plt.figure(figsize=(15,15))
plt.imshow(fallguys_wc)
plt.axis("off");
```

'Friend' and 'fun' make an appearance again for Fall Guys - in general, there is a higher proportion of positive words here.
