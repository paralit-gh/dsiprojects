#### 6. Fix Incorrect Data Types
Based on what you discovered above, use appropriate methods to re-type incorrectly typed data.
- Define a function that will allow you to convert participation rates to an appropriate numeric type. Use `map` or `apply` to change these columns in each dataframe.

````python
def part_to_numeric(part_rate):
    """ This function takes in the participation rate of each state as an argument,
        strips any trailing '%' characters and converts it to type int.
    """
    try:
        return int(part_rate.strip('%'))
    except:
        pass
````

````python
# remove the '%' symbol and convert the SAT 2017 'Participation' column to be of type int

sat_2017['Participation'] = sat_2017['Participation'].map(part_to_numeric)
````

````python
# remove the '%' symbol and convert the ACT 2017 'Participation' column to be of type int

act_2017['Participation'] = act_2017['Participation'].map(part_to_numeric)
````

#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
def standard_dev_col(col):
    """ This function takes in a 1-D data array 'col' and calculates the population standard deviation of the data 
    using standard Python functions.
    """    
    
    col_mean = sum(col)/len(col)
    variance = sum([(x - col_mean)**2 for x in col]) / len(col)
    
    return variance ** 0.5
````

- Use a **dictionary comprehension** to apply your standard deviation function to each numeric column in the dataframe.  **No loops**  
- Assign the output to variable `sd` as a dictionary where: 
    - Each column name is now a key 
    - That standard deviation of the column is the value 
     
*Example Output :* `{'ACT_Math': 120, 'ACT_Reading': 120, ...}`

````python
# use a dictionary comprehension to apply the standard_dev_col function to each column in the dataframe, and assign to variable sd

sd = {col: round(standard_dev_col(sat_act[col]), 3) for col in sat_act.columns if col != 'state'}

sd
````

````python
# calculate the standard deviation using numpy's std method

sd_numpy = {col: round(np.std(sat_act[col]), 3) for col in sat_act.columns if col != 'state'}

sd_numpy
````

````python
# calculate the standard deviation using DataFrame.std()

sd_pandas = round(sat_act.std(), 3).to_dict()

sd_pandas
````

````python
# merge all three into a DataFrame for easy comparison

pd.DataFrame(data=[sd, sd_numpy, sd_pandas], index=['sd_manual', 'sd_numpy', 'sd_pandas']).T
````

[Helpful Link for Plotting Multiple Figures](https://matplotlib.org/users/pyplot_tutorial.html#working-with-multiple-figures-and-axes)

Here's some starter code:

````python
def subplot_histograms(dataframe, list_of_columns, list_of_titles, list_of_xlabels, bins=20, binrange=None,
                       sharex=True, sharey=True):
    """ This function plots histograms for a list of columns that are part of a DataFrame with appropriate titles and 
    xlabels, and also plots the mean and mode as vertical lines for each distribution for easy visualization.
    """    
    
    nrows = int(np.ceil(len(list_of_columns)/2))
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize=(15, nrows*6), sharex=sharex, sharey=sharey)
    ax = ax.ravel()    
    fig.tight_layout(pad=3, h_pad=5)
                      
    for i, column in enumerate(list_of_columns): 
        sns.histplot(dataframe[column], ax=ax[i], binrange=binrange, bins=bins) 
        ax[i].set_title(list_of_titles[i], fontsize=20)
        ax[i].set_xlabel(list_of_xlabels[i], fontsize=15)
        ax[i].axvline(dataframe[column].mean(), color='darkturquoise', linestyle='solid', linewidth=1) 
        ax[i].axvline(dataframe[column].median(), color='midnightblue', linestyle='dotted', linewidth=1)
        ax[i].text(0.025, 0.925, 'Mean: {:.1f}'.format(dataframe[column].mean()), color='darkturquoise', transform=ax[i].transAxes)
        ax[i].text(0.025, 0.875, 'Median: {:.1f}'.format(dataframe[column].median()), color='midnightblue', transform=ax[i].transAxes)
                
    plt.show()
````

#### Plot and interpret boxplots

For each numeric variable in the dataframe create a boxplot using Seaborn. Boxplots demonstrate central tendency and spread in variables. In a certain sense, these are somewhat redundant with histograms, but you may be better able to identify clear outliers or differences in IQR, etc.

Multiple values can be plotted to a single boxplot as long as they are of the same relative scale (meaning they have similar min/max values).

Each boxplot should:

- Have clear labels for each variable
- Have appropriate titles and labels

````python
def plot_boxplots(dataframe, list_of_columns, title, xlabel, ylabel, figsize=(10, 10)):
    """ This function plots box plots for a given list of features of the dataframe.
    """    
    
    fig = plt.figure(figsize=figsize)   
    
    sns.boxplot(data=dataframe[list_of_columns], width=0.5)
        
    plt.title(title, fontsize=20)
    plt.xlabel(xlabel, fontsize=15)
    plt.ylabel(ylabel, fontsize=15)
````

#### Distributions in the data

In this dataset, each data represents a sample from a population.                        
For example, for ACT math test:
- Population: the test results of all the students who take this test, nation-wide.
- Population mean: is the national average of ACT math test (total scores/total no. of test takers) 
- Sample: the state means of ACT math test. We have 51 samples (51 states)

***According to CLT, we generally assuming that data we sample from a population will be normally distributed. Do we observe this trend?***

> **Answer:**

Based on the histograms plotted earlier for the SAT and ACT subject scores, the sampling distributions of the sample means (in this case, the scores for each state) do not appear to be normally distributed. Most of the distributions are right skewed and there are some cases where the distribution appears to be bimodal instead (e.g. ACT 18 Math and ACT 18 Reading).

In particular, the Participation sampling distributions for both the ACT and SAT in 2017 and 2018 do not resemble normal distributions at all.

Does This Assumption Hold for:

Explain your answers for each distribution and how you think this will affect estimates made from these data.

> **Answer:**

The data that is available in the datasets already comprise the entire population data, where the samples are obtained via cluster sampling, and each cluster is based on geographical boundaries (i.e. the different states). In this case, all the clusters have been selected for sampling as we have complete data for all 51 states.

For the Central Limited Thereom to be applicable such that the sampling distribution of the sample means approximates a normal distribution, the samples should be equally sized, and need to be randomly drawn (with replacement) from the population, so that each sample can be an unbiased representation of the population. In this case, the samples are already drawn based on pre-defined geographical boundaries, and the sample size of each sample is different (as each state has a different number of students who take the ACT/SAT). As such, the Central Limit Theorem would not hold, as evidenced by how the sampling distributions for *Math Score*, *Reading Score* and *Participation Rates* do not approximate normal distributions. Our sample frames already comprise the entire population set.

Estimates of the population parameters (true national average scores for each test or the participation rate) made from the sampling data may be inaccurate given that the CLT is not applicable in this case and the normality assumption does not hold.
