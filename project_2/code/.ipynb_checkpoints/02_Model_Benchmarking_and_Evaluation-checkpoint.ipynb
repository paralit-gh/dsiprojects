# Train/Test Splitting of Data

We will now prepare our features matrix (X) and target vector (y), as well as split our data into train and test sets for model evaluation.

```python
# read cleaned .csv file from previous notebook

ames = pd.read_csv('../datasets/cleaned.csv')
```

```python
# create X and y variables

X = ames.drop(columns=['SalePrice'], axis=1)
y = ames['SalePrice']
```

```python
# assign feature names to feature_names for later use

feature_names = X.columns
```

```python
# check shape of X, y

print(f'Shape of features matrix X: {X.shape}')
print(f'Shape of target vector y: {y.shape}')
```

```python
# split our training and testing data using the (default) test size of 0.25 and set random_state of 42 for reproducibility

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42) 
```

# Model Creation

We will first establish our baseline score as a reference point. This is the Root Mean Squared Error (RMSE) obtained by simply passing the mean `SalePrice` of the training set as the predicted value for our test set. 

In other words, as it is not using any of the predictor variables to predict the target variable at all, any production model that we eventually develop should be able to outperform this baseline score. It would also serve as an indication of whether we are improving our approach to the modelling process. 

## Calculate baseline score

```python
# obtain the mean of y_train

y_train_mean = y_train.mean()
```

```python
# calculate RMSE using y_train_mean by creating a 2D array 

mean_squared_error(y_test, np.full((len(y_test), 1), y_train_mean), squared=False)                 
```

**Observations**

Our baseline RMSE score is $79511.73$.

```python
# build function to calculate and display train, test, cross_val_score R2 and RMSE values for a given regressor

def train_test_cv_scores(model_type, X_train, X_test, y_train, y_test, 
                         ridge_alpha=1.0, lasso_alpha=1.0, elasticnet_alpha=1.0,
                         l1_ratio=0.5):
    """
    For a given model, training and test datasets, fit and score the model and display the corresponding metrics for
    each model (R2 and RMSE).
    """
    # build corresponding model based on model_type passed into function
    
    if model_type == 'lr':
        model = LinearRegression()
    elif model_type == 'ridge':
        model = Ridge(alpha=ridge_alpha)
    elif model_type == 'lasso':
        model = Lasso(alpha=lasso_alpha)
    elif model_type == 'elasticnet':
        model = ElasticNet(alpha=elasticnet_alpha, l1_ratio=l1_ratio)
        
    model.fit(X_train, y_train) # fit our model on training data    
    
    # obtain R2 scores
    model_train_score = model.score(X_train, y_train)    
    model_test_score = model.score(X_test, y_test)
    model_cv_score = cross_val_score(model, X_train, y_train, cv=5).mean()
    
    # obtain RMSE scores
    model_train_rmse_score = mean_squared_error(y_train, model.predict(X_train), squared=False)
    model_test_rmse_score = mean_squared_error(y_test, model.predict(X_test), squared=False)
    model_cv_rmse_score = -cross_val_score(model, X_train, y_train, scoring='neg_root_mean_squared_error', cv=5).mean()
                                               
    # print out tabulated scores
    print(f'{model_type.capitalize()} Model')
    print('--------')
    print('R2 Score')
    print('--------')
    print(f'Model Train Score: {model_train_score:.4f}')
    print(f'Model Test Score: {model_test_score:.4f}')
    print(f'Model 5-fold CV Score: {model_cv_score:.4f}')
    print('--------')
    print('RMSE')
    print('--------')
    print(f'Model Train Score: {model_train_rmse_score:.4f}')
    print(f'Model Test Score: {model_test_rmse_score:.4f}')
    print(f'Model 5-fold CV Score: {model_cv_rmse_score:.4f}')
    
    # return the coefficients of the model
    return model.coef_
```

## Scaling

To use our Ridge and Lasso regularization models, we are required to first standardize all predictor columns, including the dummy variables. Otherwise, certain features would be penalized more simply because of their scale.

This will be achieved using `StandardScaler()`.

```python
# fit StandardScaler on training data, and transform both training and test data using the same StandardScaler

ss = StandardScaler()
ss.fit(X_train)
X_train_scaled = ss.transform(X_train)
X_test_scaled = ss.transform(X_test)
```

## Linear Regression

```python
# print out scores for LR model and return coefficients

linear_coefficients = train_test_cv_scores('lr', X_train_scaled, X_test_scaled, y_train, y_test)
```

**Observations**

We note that the linear model that we fitted produces rather abysmal cross-validated results - the R^2 is highly negative ($-1.87 \times 10^{25}$) and the RMSE is of magnitude $10^{17}$. This is considerably worse than our baseline score.

As the cross-validated score provides us an estimate of the model's performance on unseen data, this is somewhat within expectations, given the sheer number of features that were used to train the model (119) - we would expect a high degree of overfitting. This can be corroborated by scoring the model on the training and test data.

Indeed, the LinearRegression model on the training data gives us an excellent $R^2$ score of $0.895$ and RMSE of $25655.56$

However, this model grossly overfits the training data as the test data yields $R^2$ score of $-5.54 \times 10^{22}$ and RMSE of $1.8714 \times 10^{16}$.

```python
# visualize coefficients in a dataframe

linear_coef_df = pd.DataFrame(data={'Feature': feature_names, 'Coefficient': linear_coefficients}).sort_values(by='Coefficient', ascending=False, key=abs)
linear_coef_df.head(10)
```

**Observations**

By examining the coefficients, we see that some of these are of extremely large magnitude, which are well beyond the range of reasonable values.

Hence, we will attempt to reduce the number of features by using some *Embedded Methods* of feature selection: Lasso, Ridge and ElasticNet. These regularization techniques help to constrain and shrink the coefficients, shrinking them towards 0. This helps to reduce our model variance and address the issue of overfitting.

In particular, Lasso will help to reduce some of the coefficients to 0, eliminating them entirely.

## Recursive Feature Elimination

Recursive Feature Elimination (RFE) is a wrapper method for feature selection that recursively removes attributes and builds subsequent models on those attributes that remain. We will pass our training dataset with 63 features through RFE to obtain our desired number of 30 features.

As RFE is a computationally expensive procedure, it is undesirable to pass the entire full feature set into it right from the start, especially if only one feature is being removed during each iteration. Hence, we have left it as our last means of performing feature elimination.

```python
# create instance of RFE

n_features_to_select = 30

lr = LinearRegression()
rfe = RFE(lr, n_features_to_select=n_features_to_select, verbose=1)
rfe.fit(X_train_scaled, y_train)
```

```python
# determine which features are being kept

rfe.support_
```

```python
# display ranking of features

rfe.ranking_
```

```python
# show names of features that are being kept

rfe_features = lasso_kept_features[rfe.support_]
rfe_features
```

```python
# display ranking and feature names in a dataframe

rfe_ranking_df = pd.DataFrame(data={'Feature': lasso_kept_features, 'Ranking': rfe.ranking_})
rfe_ranking_df
```

**Observations**

RFE has successfully reduced our number of features to 30. We will pass those features with rank 1 (i.e. those that are retained) to the linear models again.

## Hyperparameter Tuning

For our `Lasso` model, the hyperparameter associated with it is the $\alpha$ value, which also serves as the tuning parameter. By increasing the $\alpha$ value, we increase the amount of regularization, which would in turn increase the bias and reduce the variance of our model. 

Earlier on, the `LassoCV` object had helped to search through varying $\alpha$ values to find the optimal value, leading to the lowest RMSE.

We will visualize the effect of varying the hyperparameter $\alpha$ to ascertain that this is indeed the case.

```python
# adapted plot_cv function to plot the average MSE score obtained from cross-validation for each value of alpha

def plot_cv(alphas, cv_means, optimal_alpha):

    plt.figure(figsize=(12,8))
    plt.plot(alphas, cv_means, lw=2)
    plt.axvline(optimal_alpha, linestyle='dashed', color='darkblue')
    plt.xlabel('alpha')
    plt.ylabel('Mean Squared Error')
    plt.title('Plot of MSE against alpha', fontsize=20)
```

```python
# instantiate a LassoCV object

l_alphas = np.arange(1, 100, 1)
lasso_cv = LassoCV(alphas=l_alphas, cv=5)
lasso_cv.fit(X_train_scaled, y_train)
```

```python
# obtain optimal alpha value

optimal_lasso_alpha = lasso_cv.alpha_
optimal_lasso_alpha
```

```python
lasso_cv_means = [np.mean(cv_alpha) for cv_alpha in lasso_cv.mse_path_]

plot_cv(lasso_cv.alphas_, lasso_cv_means, optimal_lasso_alpha);
```

**Observations**

Based on the above plot of RMSE against $\alpha$ for the `LassoCV` model, we see that `LassoCV` has indeed helped to select the value of $\alpha$ that minimized the Mean Square Error across the range of $\alpha$ specified. 

When we earlier specified `n_alphas = 100` instead of an actual range for `alpha` while building the `LassoCV` model, it helped to create a grid of $\alpha$ values such that `alpha_min / alpha_max = 1e-3`, which is the default value for `eps`. This has worked well enough for our case, given that the magnitude of change for the MSE between $\alpha$ values of 0 to 100 is extremely miniscule and insignificant.

As such, we would not be performing any further hyperparameter tuning for our model.

# Production Model

We have chosen our `Lasso` model with 30 features to be our production model. The relatively low number of features (compared to the original 119) allows for the model to be more interpretable and generalizable to unseen data. However, there is the possibility that the bias may be higher as compared to if there were more features (e.g. around 60).

```python
# train our final Lasso model on the training data

final_lasso = Lasso(alpha=optimal_lasso_alpha)
final_lasso.fit(X_train_scaled, y_train)
```

```python
# print out scores for Lasso model and return coefficients

lasso_coefficients = train_test_cv_scores('lasso', X_train_scaled, X_test_scaled, y_train, y_test, lasso_alpha=optimal_lasso_alpha)
```

We will focus on the RMSE value instead of the $R^2$ or adjusted $R^2$ value as our evaluation metric, as it directly measures the average error of our predictions against the actual sale price, and is a more interpretable metric.

**Observations**

Our model performs reasonably well as it can be seen that a majority of predicted prices coincide roughly with the actual price, forming a relatively straight line. There is one particular point where the model undervalued the `SalePrice` by almost 150k. Given that a large proportion of actual `SalePrice` were in the range of 100k to 200k, it is possible that our model is unable to generalize well to `SalePrice` that is well beyond this range.

Indeed, after about actual `SalePrice` of around 350k, our model consistently under predicts most of the prices of houses in this range. It could be possible that our model had failed to capture and account for some features that can adequately explain the prices of higher-end houses.

## Model Limitations

When we use multiple linear regression to model the variables, we are inherently making 5 assumptions:

- Linearity
- Independence of Error
- Normality
- Equality of Variance
- Independence of Predictors

However, it may not be the case that house prices vary linearly with our predictor variables. For instance, especially large houses (with large `TotalLivArea`) might command a premium, and may be sold at a disproportionately higher price.

Our errors are also normally distributed for the most part, there are some outliers that may skew the model during training and affect its performance on unseen data. The model consistently under-predicts the `SalePrice` of more expensive houses. It can also be seen that the residuals tend to be more sparsely distributed as the predicted price increases. This suggests that the equality of variance condition is not fully met and the errors are not homoscedastic.

Finally, we have tried to eliminate predictor features that tend to be highly collinear based on their Pearson correlation coefficient, and hence Independence of Predictors should be satisfied to a certain extent.

Given that a number of the above assumptions may not be fully satisfied, our linear regression model's performance may be impacted to some extent.

In addition, the dataset is limited to home sales made in Ames, Iowa from 2006 to 2010. It is unlikely that the model can be well extrapolated to today's context, due to various factors such as inflation and different market sentiments.

## Further Refinements to Model

To further improve the model, these are some of the steps that could be undertaken:

- Perform more research on predictor variables to understand their potential effect on `SalePrice` to allow us to evaluate model better
- Do outside research on Ames housing and bring in more relevant predictor variables to be used for modelling
- Use other machine learning algorithms such as `DecisionTreeRegressor` for RFE to do feature selection
- As our target variable `SalePrice` exhibits a slight right skew (with a large cluster of homes around the mean/median price and relatively fewer expensive homes), this could lead to our model making biased predictions for homes with larger prices. We could consider performing a logarithmic transformation on our target variable first to normalize the data, prior to performing our modelling. This would also require us to back-calculate the $R^2$ and RMSE values after transforming the target Y variables with an exponential function first.
