```python
def plot_boxandhist(dataframe, column_name, categorical_col, figsize=(10, 14)):
    """ This function plots box plots for a given column name, split by the various categories of another column of the dataframe.
        It also plots the histogram for the categorical column.
    """    
    
    fig, ax = plt.subplots(nrows=2, figsize=figsize)   
    
    sns.boxplot(y=dataframe[column_name], x=dataframe[categorical_col], width=0.5, ax=ax[0])
    sns.countplot(x=dataframe[categorical_col], linewidth=1.5, edgecolor='black', ax=ax[1])

    ax[0].set_title(f'Boxplot of {column_name} against {categorical_col}', fontsize=20)
    ax[0].set_xlabel(categorical_col, fontsize=15)
    ax[0].set_ylabel(column_name, fontsize=15)
    ax[1].set_title(f'Histogram of {categorical_col}', fontsize=20)
    ax[1].set_xlabel(categorical_col, fontsize=15)
    ax[1].set_ylabel('Count', fontsize=15)
```

## Feature engineering, encoding

```python
ames_dropped.dtypes
```

```python
ames_dropped.select_dtypes(include='int64').head()
```

```python
ames_dropped.select_dtypes(include='float64').head()
```

```python
ames_dropped.select_dtypes(include='object').head()
```

```python
# get dummies

ames_dummies = pd.get_dummies(ames_dropped, columns=ames_dropped.select_dtypes(include=['category', 'object']).columns, drop_first=False)
```

```python
ames_dummies.shape
```

```python
ames_dummies.head()
```

```python
ames_dummies.dtypes
```

```python
# Model Prep: Create our features matrix (X) and target vector (y)
features = [col for col in ames_dummies._get_numeric_data() if col != 'SalePrice']
X = ames_dummies[features] # makes sure 'Apps' is not in X - data leakage
y = ames_dummies['SalePrice']
```

```python
X.head()
```

```python
X.shape
```

```python
y.shape
```

```python
# split our training and testing data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42) # remember to set random state!
```

```python
# scale and fit

ss = StandardScaler()
ss.fit(X_train)
X_train_scaled = ss.transform(X_train)
X_test_scaled = ss.transform(X_test)
```

# Build first model

## Linear Regression

### Linear Regression, evaluation using CV

```python
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)
```

```python
# LR Estimate of Test R2 using cross validation
# by default, regression metric is R2
lr_cv_scores = cross_val_score(lr, X_train_scaled, y_train, cv=5)
lr_cv_scores.mean()
```

```python
# LR Estimate of Test RMSE using cross validation
lr_cv_rmse_scores = -cross_val_score(lr, X_train_scaled, y_train, scoring='neg_root_mean_squared_error', cv=5)
lr_cv_rmse_scores.mean()
```

```python
lr.intercept_
```

```python
pd.DataFrame(data=[lr.coef_], columns=features, index=['Coefficient Score']).T.sort_values(by='Coefficient Score', ascending=False, key=abs)

# there are some gigantic coefficients e.g. GarageType_None, BsmtQual_None, ExterQual_Po, KitchenQual_Po
```

### Linear Regression, test evaluation

```python
print('Training R2 (LR): ' + str(lr.score(X_train_scaled, y_train)))
print('Estimate of Testing R2 (LR): ' + str(lr_cv_scores.mean()))
print('Test R2 (LR) ' + str(lr.score(X_test_scaled, y_test)))
```

```python
print('Training RMSE (LR): ' + str(mean_squared_error(y_train, lr.predict(X_train_scaled), squared=False)))
print('Estimate of Testing MSE (LR): ' + str(lr_cv_rmse_scores.mean()))
print('Test RMSE (LR): ' + str(mean_squared_error(y_test, lr.predict(X_test_scaled), squared=False)))
```

## Ridge Regression

### Ridge Regression, evaluation using CV

```python
# Ridge Model
ridge = RidgeCV(alphas = np.logspace(0, 5, 200)) # in the first place, the range must be suitable, otherwise RidgeCV cannot output a good value
ridge.fit(X_train_scaled, y_train)
```

```python
ridge.alpha_ # our best alpha value
```

```python
# Ridge Estimate of Test R2 using cross validation
ridge_cv_scores = cross_val_score(ridge, X_train_scaled, y_train, cv=5)
ridge_cv_scores.mean()
```

```python
# Ridge Estimate of Test MSE using cross validation
ridge_cv_rmse_scores = -cross_val_score(ridge, X_train_scaled, y_train, scoring='neg_root_mean_squared_error', cv=5) 
ridge_cv_rmse_scores.mean()
```

### Ridge Regression, test evaluation

```python
print('Training R2 (Ridge): ' + str(ridge.score(X_train_scaled, y_train)))
print('Estimate of Testing R2 (Ridge): ' + str(ridge_cv_scores.mean()))
print('Test R2 (Ridge) ' + str(ridge.score(X_test_scaled, y_test)))
```

```python
print('Training RMSE (Ridge): ' + str(mean_squared_error(y_train, ridge.predict(X_train_scaled), squared=False)))
print('Estimate of Testing RMSE (Ridge): ' + str(ridge_cv_rmse_scores.mean()))
print('Test RMSE (Ridge): ' + str(mean_squared_error(y_test, ridge.predict(X_test_scaled), squared=False)))
```

```python
pd.DataFrame(data=[ridge.coef_], columns=features, index=['Coefficient Score']).T.sort_values(by='Coefficient Score', ascending=False, key=abs)
```
